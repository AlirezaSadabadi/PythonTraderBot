```python
import pandas as pd
import yfinance as yf

# دانلود کندل ها
df_raw = yf.download(["SPY", "UPRO"], start="2023-01-01", end="2025-07-28",interval='1d')
df_raw
```

```python
# جداسازی اطلاعات
df_upro = df_raw.xs('UPRO', level=1, axis=1)
df_spy = df_raw.xs('SPY',  level=1, axis=1)
df_spy
```

```python
# عوض کردن اسم ستون ها
df_spy.columns = ['open_SPY', 'high_SPY', 'low_SPY', 'close_SPY', 'volume_SPY']

# تلفیق مجدد اطلاعات
df_merged = pd.concat([df_spy, df_upro], axis=1)
df_merged.dropna(inplace=True)

df_merged.reset_index(inplace=True)
df_merged
```

```python
from backtesting import Backtest, Strategy
from backtesting.test import SMA

class LeveragedLongRun(Strategy):
    SMA_Length = 100

    def init(self):
        # محاسبه میانگین متحرک
        self.sma = self.I(SMA, self.data.df['close_SPY'], self.SMA_Length)

    def next(self):
        spy_close = self.data.df['close_SPY'].iloc[self.data.index[-1]] # آخرین مقدار
        sma_value = self.sma[-1]  # آخرین مقدار

        # در صورتی که کلوز کندل بالای میانگین متحرک رفت خرید بزن
        if spy_close > sma_value:
            if not self.position:
                self.buy()
        # در غیر اینصورت ببندش
        else:
            if self.position:
                self.position.close()
```

```python
# اجرای بک تست
backtest = Backtest(
    df_merged,
    LeveragedLongRun,
    cash=100000,
    commission=0.0002,
    trade_on_close=False
)
result = backtest.run()
print(result)
```
