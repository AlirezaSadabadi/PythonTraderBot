```python
import pandas as pd
import yfinance as yf

# لیستی از ۲۰ سهام بازار آمریکا
tickers = [
    'GOOGL', 'AMZN', 'META', 'AAPL', 'MSFT', 
    'TSLA', 'NVDA', 'CEG', 'JNJ', 'V',
    'SPY', 'QQQ', 'DIA', 'IWM', 'ARKK',
    'DG', 'XLK', 'XLV', 'XLY', 'CRWV'
]

# دانلود ۱۰ سال و نیم داده در تایم فریم یک روزه
data = yf.download(tickers, start="2015-01-01", end="2025-06-04", group_by='ticker', interval='1d', auto_adjust=False)
```

```python
# تابع محاسبه معیار
def count_candle_similar(n, data):
    downSimilar = 0
    buy_followed_by_green = 0
    upSimilar = 0
    sell_followed_by_red = 0

    for ticker in tickers:
        df = data[ticker][['Open', 'Close']].dropna()
        # True: green, False: red
        candleColor = df['Close'] > df['Open']  
        candles = candleColor.values

        i = 0
        while i < len(candles) - n:
            # چک کردن کندل های سبز پشت سرهم
            if all(candles[i + j] for j in range(n)):
                upSimilar += 1
                if i + n < len(candles) and not candles[i + n]:
                    sell_followed_by_red += 1
                i += n
            # چک کردن کندل های قرمز پشت سرهم
            elif all(not candles[i + j] for j in range(n)):
                downSimilar += 1
                if i + n < len(candles) and candles[i + n]:
                    buy_followed_by_green += 1
                i += n
            else:
                i += 1

    upFraction = sell_followed_by_red / upSimilar if upSimilar > 0 else 0
    downFraction = buy_followed_by_green / downSimilar if downSimilar > 0 else 0

    return {
        f"{n} Red Candles Followed by Green (Buy Position)": downFraction,
        f"{n} Green Candles Followed by Red (Sell Position)": upFraction
    }
```

```python
metrics = count_candle_similar(2, data)
print(metrics)
```

```python
from backtesting import Strategy

class ConsecutiveBuySellStrategy(Strategy):
    # تعداد کندل مشابه پشت سر هم
    n = 2

    def init(self):
        pass

    def next(self):
        # حداقل ان کندل باید باشد تا بتوانیم شروع کنیم
        if len(self.data) < self.n:
            return

        open_ = self.data.Open
        close_ = self.data.Close

        # بعد از تعداد مشخص شده از کندل قرمز یک کندل سبز دیده می شود و پوزیشن خرید باز می شود
        if all(close_[-i] < open_[-i] for i in range(1, self.n + 1)):
            self.buy()

        # بعد از تعداد مشخص شده از کندل سبز یک کندل قرمز دیده می شود و پوزیشن فروش باز می شود
        elif all(close_[-i] > open_[-i] for i in range(1, self.n + 1)):
            self.sell()

        # بستن یک پوزیشن با پایان یافتن کندل
        # البته در کندل بعدی پوزیشن قبلی بسته می شود
        # اما به هر حال ما به هدفمان می رسیم
        if self.position:
            self.position.close()
```

```python
from backtesting import Strategy, Backtest

class ConsecutiveBuyStrategy(Strategy):
    # تعداد کندل مشابه پشت سر هم
    n = 4

    def init(self):
        pass

    def next(self):
        # حداقل ان کندل باید باشد تا بتوانیم شروع کنیم
        if len(self.data) < self.n:
            return


        if all(self.data.Close[-i] < self.data.Open[-i] for i in range(1, self.n + 1)):
            self.buy()

        if self.position:
            self.position.close()
```

```python
import numpy as np

results = {}

for symbol in tickers:
    df = data[symbol][['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
    df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

    backtest = Backtest(df,
                  ConsecutiveBuySellStrategy,
                  cash=1000,
                  commission=0.0002, # for crypto 0.001 or 0.002, for futures 0.0002, for forex 0.0001
                  exclusive_orders=True,
                  margin=1/1)

    stats = backtest.run()
    results[symbol] = stats

# استخراج اطلاعات نماد و مقدار سود برگشتی به تفکیک نمادها در قالب یک لیست
returns_list = [(sym, res['Return [%]']) for sym, res in results.items()]

print("--Returns for each ticker--")
for sym, ret in returns_list:
    print(f"{sym}: {ret:.2f}%")

# محاسبه معیار تجمیعی
sum_of_returns = sum(ret for _, ret in returns_list)
average_return = np.mean([ret for _, ret in returns_list])
max_drawdowns = [(sym, res['Max. Drawdown [%]']) for sym, res in results.items()]
max_drawdown_overall = min(mdd for _, mdd in max_drawdowns)

# چاپ معیارهای تجمیعی
print("\n-- Aggregated Metrics --")
print(f"Sum of Returns: {sum_of_returns:.2f}%")
print(f"Average Return: {average_return:.2f}%")
print(f"Maximum Drawdown across all tickers: {max_drawdown_overall:.2f}%")

# ذخیره در قالب یک دیکشنری برای بهره برداری های آتی
summary_metrics = {
    'Sum of Returns': sum_of_returns,
    'Average Return': average_return,
    'Max Drawdown Overall': max_drawdown_overall
}
```
